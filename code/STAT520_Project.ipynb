```python
# ----------- CHOOSE THE STAT TO COMPARE -----------------
nfl_col = "nfl-y/a"     # NFL dataset column name
college_col = "Y/A"     # College dataset column name
# ---------------------------------------------------------

# Convert columns to numeric and drop missing
nfl_metric = pd.to_numeric(nfl[nfl_col], errors="coerce").dropna()
college_metric = pd.to_numeric(collegeQB[college_col], errors="coerce").dropna()

print("NFL sample size:", len(nfl_metric))
print("College sample size:", len(college_metric))
```

```python
# ----------------- Mann-Whitney U test -------------------
u_stat, u_p = stats.mannwhitneyu(
    nfl_metric,
    college_metric,
    alternative="two-sided"
)
print("\nMannâ€“Whitney U Test")
print("U statistic:", u_stat)
print("p-value:", u_p)
```

```python
# ---------------- Kolmogorovâ€“Smirnov test ----------------
ks_stat, ks_p = stats.ks_2samp(nfl_metric, college_metric)
print("\nKS Test")
print("KS statistic:", ks_stat)
print("p-value:", ks_p)
```

```python
# ------------------- Permutation test ---------------------
obs_diff = nfl_metric.mean() - college_metric.mean()

combined = np.concatenate([nfl_metric.values, college_metric.values])
n_nfl = len(nfl_metric)
n_perm = 10000

count = 0
for _ in range(n_perm):
    np.random.shuffle(combined)
    fake_nfl = combined[:n_nfl]
    fake_college = combined[n_nfl:]
    diff = fake_nfl.mean() - fake_college.mean()
    if abs(diff) >= abs(obs_diff):
        count += 1

perm_p = count / n_perm
print("\nPermutation Test")
print("Observed difference in means:", obs_diff)
print("Permutation p-value:", perm_p)
```

```python
# qb_draft_project.py

import pandas as pd
import numpy as np
from pathlib import Path

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# =========================================================
# 1. Load data
# =========================================================

DATA_PATH = Path("/content/drive/MyDrive/STAT520/QB_ALL_TableSchema.csv")

df = pd.read_csv(DATA_PATH)

# Quick sanity check
print(df.head())
print(df.dtypes)

```

```python

# =========================================================
# 2. Basic cleaning and filtering
#    - Keep only historical college QBs (not current pros)
#    - Focus on last 10 draft classes with known drafted flag
# =========================================================

print("DRAFTED_FLAG distribution after initial load:")
print(df["DRAFTED_FLAG"].value_counts(dropna=False))

# Keep rows that have a drafted flag (0/1) and some QB stats
df = df[df["DRAFTED_FLAG"].isin([0, 1])].copy()
print("\nDRAFTED_FLAG distribution after filtering for 0 or 1:")
print(df["DRAFTED_FLAG"].value_counts(dropna=False))

# Some rows are PROSPECTS with blank draft year; cast to numeric safely
df["YEAR_DRAFTED"] = pd.to_numeric(df["YEAR_DRAFTED"], errors="coerce")

# Keep only rows with a valid draft year
df = df[~df["YEAR_DRAFTED"].isna()].copy()
print("\nDRAFTED_FLAG distribution after filtering for valid YEAR_DRAFTED:")
print(df["DRAFTED_FLAG"].value_counts(dropna=False))

# Focus on last 10 draft years present in the file
last_10_years = (
    df["YEAR_DRAFTED"]
    .dropna()
    .sort_values()
    .unique()[-10:]
)
df = df[df["YEAR_DRAFTED"].isin(last_10_years)].copy()

print("Draft years used:", sorted(df["YEAR_DRAFTED"].unique()))
print("Number of QBs:", len(df))
print("\nDRAFTED_FLAG distribution after filtering for last 10 years:")
print(df["DRAFTED_FLAG"].value_counts(dropna=False))
```

```python

# =========================================================
# 3. Feature selection / engineering
#    We use physical attributes + key college stats.
#    You can adjust this list depending on your analysis goals.
# =========================================================

# Example numeric features (present in your schema)
numeric_features = [
    "HEIGHT_IN",
    "WEIGHT_LBS",
    "CMP",
    "ATT",
    "CMP_PCT",
    "YDS",
    "TD",
    "INTS",
    "Y_PER_ATT",
    "ADJ_Y_PER_ATT",
    "RATE",
    "R_ATT",
    "R_YDS",
    "R_AVG",
    "R_TDS",
]

# Some rows may have all-NaN stats (e.g., depth QBs); drop them
df_numeric = df[numeric_features]
df = df[~df_numeric.isna().all(axis=1)].copy()

# Target
TARGET = "DRAFTED_FLAG"

```

```python

# =========================================================
# 4. Train / test split
# =========================================================

# --- Diagnostic check for DRAFTED_FLAG distribution ---
print("Distribution of DRAFTED_FLAG in df before train/test split:")
print(df[TARGET].value_counts())
print("--------------------------------------------------")

X = df[numeric_features]
y = df[TARGET].astype(int)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)
```

```python
# =========================================================
# 5. Define preprocessing and models
# =========================================================

# All chosen features are numeric; scale them
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
    ]
)

# Logistic Regression pipeline (interpretable baseline)
log_reg_clf = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("model", LogisticRegression(max_iter=500, class_weight="balanced")),
    ]
)

# Random Forest pipeline (nonlinear, handles interactions)
rf_clf = Pipeline(
    steps=[
        ("preprocess", preprocessor),
        ("model", RandomForestClassifier(
            n_estimators=300,
            random_state=42,
            class_weight="balanced_subsample",
        )),
    ]
)
```

```python
# =========================================================
# 6. Train and evaluate models
# =========================================================

def evaluate_model(name, model, X_tr, y_tr, X_te, y_te):
    print(f"\n=== {name} ===")

    # Add a check for the number of unique classes in y_tr
    unique_classes_tr = y_tr.nunique()
    print(f"Training target class distribution:\n{y_tr.value_counts()}")
    if unique_classes_tr < 2:
        print(f"Error: Cannot train {name}. The training data (y_tr) contains only {unique_classes_tr} unique class(es).")
        print("Please ensure your training data contains at least two classes (e.g., both drafted and undrafted players).")
        return None # Return None or raise an exception to stop further execution for this model

    model.fit(X_tr, y_tr)

    # Cross-validated ROC-AUC on training set for a quick sense of generalization
    cv_scores = cross_val_score(
        model, X_tr, y_tr, cv=5, scoring="roc_auc"
    )
    print("CV ROC-AUC (mean \u00b1 std): "
          f"{cv_scores.mean():.3f} \u00b1 {cv_scores.std():.3f}")

    # Test-set performance
    y_pred = model.predict(X_te)
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_te)[:, 1]
        print("Test ROC-AUC:", roc_auc_score(y_te, y_proba))
    print("Classification report:")
    print(classification_report(y_te, y_pred))

    return model


log_reg_model = evaluate_model(
    "Logistic Regression", log_reg_clf, X_train, y_train, X_test, y_test
)

rf_model = evaluate_model(
    "Random Forest", rf_clf, X_train, y_train, X_test, y_test
)
```
